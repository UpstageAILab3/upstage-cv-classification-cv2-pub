```python
import pandas as pd 
import numpy as np 

import os 
from glob import glob 

from PIL import Image
import albumentations as A
from albumentations import ImageOnlyTransform
from augraphy import *

from tqdm import tqdm

pre_path = '/kkh/'
train_kr_image_path = pre_path + 'data/train_kr'
train_kr_aug_image_path = pre_path + 'data/train_kr_aug'
meta_kr_csv_path = pre_path + 'data/meta_kr.csv'
train_kr_csv_path = pre_path + 'data/train_kr.csv'
meta_kr_df = pd.read_csv(pre_path + 'data/meta_kr.csv')
train_kr_df = pd.read_csv(pre_path + 'data/train_kr.csv')

test_image_path = pre_path + 'data/test'
test_aug_image_path = pre_path + 'data/test_aug'
test_df = pd.read_csv(pre_path + 'data/sample_submission.csv')
```

```python
# train_kr_aug_image_path í´ë” ìƒì„±
try:
    os.makedirs(train_kr_aug_image_path)
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# ìˆ˜í‰ ë’¤ì§‘ê¸° ë³€í™˜ (í™•ë¥  100%)
horizontal_flip = A.Compose([
    A.HorizontalFlip(p=1),
])

# ìˆ˜ì§ ë’¤ì§‘ê¸° ë³€í™˜ (í™•ë¥  100%)
vertical_flip = A.Compose([
    A.VerticalFlip(p=1),
])

# ìˆ˜í‰ ë° ìˆ˜ì§ ë’¤ì§‘ê¸° ë³€í™˜ (í™•ë¥  100%)
double_flip = A.Compose([
    A.HorizontalFlip(p=1),
    A.VerticalFlip(p=1),
])

# ì „ì¹˜(transpose) ë³€í™˜ (í™•ë¥  100%)
transpose = A.Compose([
    A.Transpose(p=1)
])

# ì „ì¹˜ í›„ ìˆ˜í‰ ë’¤ì§‘ê¸° ë³€í™˜ (í™•ë¥  100%)
transpose_hflip = A.Compose([
    A.Transpose(p=1), 
    A.HorizontalFlip(p=1),
])

# ì „ì¹˜ í›„ ìˆ˜ì§ ë’¤ì§‘ê¸° ë³€í™˜ (í™•ë¥  100%)
transpose_vflip = A.Compose([
    A.Transpose(p=1),
    A.VerticalFlip(p=1),
])

# ì „ì¹˜ í›„ ìˆ˜í‰ ë° ìˆ˜ì§ ë’¤ì§‘ê¸° ë³€í™˜ (í™•ë¥  100%)
transpose_dflip = A.Compose([
    A.Transpose(p=1),  
    A.HorizontalFlip(p=1),
    A.VerticalFlip(p=1),
])

# íŒ¨í„´ ìƒì„±ê¸° ê°ì²´ë¥¼ ì´ˆê¸°í™”í•©ë‹ˆë‹¤.
pattern = PatternGenerator(
    imgx=120,                # ìƒì„±í•  íŒ¨í„´ ì´ë¯¸ì§€ì˜ ê°€ë¡œ í¬ê¸°
    imgy=120,                # ìƒì„±í•  íŒ¨í„´ ì´ë¯¸ì§€ì˜ ì„¸ë¡œ í¬ê¸°
    n_rotation_range=(10, 15),  # íšŒì „ ë²”ìœ„ (10ë„ì—ì„œ 15ë„ ì‚¬ì´)
    alpha_range=(0.1, 0.2),    # ì•ŒíŒŒ(íˆ¬ëª…ë„) ë²”ìœ„
    color=(130, 130, 130),    # íŒ¨í„´ì˜ ìƒ‰ìƒ (RGB)
)

# íŒ¨í„´ ìƒì„± ë³€í™˜ í´ë˜ìŠ¤ ì •ì˜
class PatternGeneratorTransform(ImageOnlyTransform):
    def __init__(self, pattern_generator, always_apply=False, p=1.0):
        super().__init__(always_apply, p)
        self.pattern_generator = pattern_generator

    def apply(self, img, **params):
        # ì´ë¯¸ì§€ì— íŒ¨í„´ ìƒì„±ê¸°ë¥¼ ì ìš©í•˜ì—¬ ë³€í™˜ëœ ì´ë¯¸ì§€ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.
        return self.pattern_generator(img)

# ë‹¤ì–‘í•œ ì´ë¯¸ì§€ ë³€í™˜ì„ ì¡°í•©í•œ ë³µí•© ë³€í™˜ ì •ì˜
noise_rotate = A.Compose([
    # 80% í™•ë¥ ë¡œ ë‹¤ìŒ ì¤‘ í•˜ë‚˜ë¥¼ ì ìš©:
    A.OneOf([
        A.GaussNoise(var_limit=(100, 900), per_channel=False, p=1),  # ê°€ìš°ì‹œì•ˆ ë…¸ì´ì¦ˆ ì¶”ê°€
        PatternGeneratorTransform(pattern, p=1)  # íŒ¨í„´ ìƒì„± ë³€í™˜
    ], p=0.8),
    
    # 50% í™•ë¥ ë¡œ ë‹¤ìŒ ì¤‘ í•˜ë‚˜ë¥¼ ì ìš©:
    A.OneOf([
        A.MotionBlur(p=0.2),                      # ëª¨ì…˜ ë¸”ëŸ¬
        A.MedianBlur(blur_limit=3, p=0.2),        # ë¯¸ë””ì–¸ ë¸”ëŸ¬
        A.Blur(blur_limit=(1, 3), p=1),           # ë¸”ëŸ¬ (1ë¶€í„° 3ê¹Œì§€ì˜ ë²”ìœ„)
    ], p=0.5),

    # ë°ê¸°ì™€ ëŒ€ë¹„ë¥¼ ì¡°ì •í•˜ëŠ” ë³€í™˜ (í™•ë¥  20%)
    A.RandomBrightnessContrast(p=0.2),

    # ìƒ‰ì¡°, ì±„ë„, ëª…ë„ë¥¼ ì¡°ì •í•˜ëŠ” ë³€í™˜ (í™•ë¥  20%)
    A.HueSaturationValue(p=0.2),

    # ì´ë¯¸ì§€ë¥¼ í‘ë°±ìœ¼ë¡œ ë³€í™˜ (í™•ë¥  30%)
    A.ToGray(p=0.3),

    # ì´ë¯¸ì§€ë¥¼ ì´ë™, ìŠ¤ì¼€ì¼, íšŒì „ ë³€í™˜ (í™•ë¥  50%)
    A.ShiftScaleRotate(
        shift_limit=0.2,      # ì´ë™ ë²”ìœ„
        scale_limit=0,        # ìŠ¤ì¼€ì¼ ë²”ìœ„ (0ì€ ìŠ¤ì¼€ì¼ ì—†ìŒ)
        rotate_limit=30,      # íšŒì „ ë²”ìœ„ (ìµœëŒ€ 30ë„)
        p=0.5,                # ë³€í™˜ ì ìš© í™•ë¥ 
        border_mode=0,        # ë³€í™˜ í›„ ê²½ê³„ ì²˜ë¦¬ ë°©ì‹ (0ì€ ìƒìˆ˜ ê°’ìœ¼ë¡œ ì±„ìš°ê¸°)
        value=[255, 255, 255] # ê²½ê³„ ì˜ì—­ì˜ ìƒìˆ˜ ê°’ (í°ìƒ‰)
    ),
])
```

```python
# ì—¬ëŸ¬ ì´ë¯¸ì§€ ë³€í™˜ì„ ì •ì˜í•œ ë¦¬ìŠ¤íŠ¸ì…ë‹ˆë‹¤. ê° ë³€í™˜ì€ íŠœí”Œë¡œ ë˜ì–´ ìˆìœ¼ë©°, íŠœí”Œì˜ ì²« ë²ˆì§¸ ìš”ì†ŒëŠ” ë³€í™˜ì˜ ì ‘ë‘ì‚¬(prefix)ì´ê³ , ë‘ ë²ˆì§¸ ìš”ì†ŒëŠ” ë³€í™˜ ê°ì²´ì…ë‹ˆë‹¤.
base_aug_types = [
    ("hf_", horizontal_flip),       # ìˆ˜í‰ ë’¤ì§‘ê¸° ë³€í™˜, ì ‘ë‘ì‚¬ "hf_"
    ("vf_", vertical_flip),         # ìˆ˜ì§ ë’¤ì§‘ê¸° ë³€í™˜, ì ‘ë‘ì‚¬ "vf_"
    ("df_", double_flip),           # ìˆ˜í‰ ë° ìˆ˜ì§ ë’¤ì§‘ê¸° ë³€í™˜, ì ‘ë‘ì‚¬ "df_"
    ("t_", transpose),              # ì „ì¹˜ ë³€í™˜, ì ‘ë‘ì‚¬ "t_"
    ("thf_", transpose_hflip),      # ì „ì¹˜ í›„ ìˆ˜í‰ ë’¤ì§‘ê¸° ë³€í™˜, ì ‘ë‘ì‚¬ "thf_"
    ("tvf_", transpose_vflip),      # ì „ì¹˜ í›„ ìˆ˜ì§ ë’¤ì§‘ê¸° ë³€í™˜, ì ‘ë‘ì‚¬ "tvf_"
    ("tdf_", transpose_dflip)       # ì „ì¹˜ í›„ ìˆ˜í‰ ë° ìˆ˜ì§ ë’¤ì§‘ê¸° ë³€í™˜, ì ‘ë‘ì‚¬ "tdf_"
]
```

```python
# ë¹ˆ ë¦¬ìŠ¤íŠ¸ë¥¼ ì´ˆê¸°í™”í•©ë‹ˆë‹¤. ë³€í™˜ëœ ì´ë¯¸ì§€ì˜ IDì™€ íƒ€ê²Ÿ ì •ë³´ë¥¼ ì €ì¥í•  ê²ƒì…ë‹ˆë‹¤.
ids = []
targets = []

# `train_kr_df` DataFrameì˜ ê° í–‰ì— ëŒ€í•´ ë°˜ë³µí•©ë‹ˆë‹¤.
# `itertuples()`ëŠ” DataFrameì„ íŠœí”Œ í˜•íƒœë¡œ ë°˜ë³µí•  ìˆ˜ ìˆê²Œ í•´ì¤ë‹ˆë‹¤.
for index, ID, target in tqdm(test_df.itertuples(), total=test_df.shape[0], desc='Image augmentation'):
    # ì´ë¯¸ì§€ì˜ íŒŒì¼ ê²½ë¡œë¥¼ ìƒì„±í•©ë‹ˆë‹¤.
    image_path = os.path.join(test_image_path, ID)
    
    # ì´ë¯¸ì§€ íŒŒì¼ì„ ì—´ê³ , numpy ë°°ì—´ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
    image = np.array(Image.open(image_path))
    
    # `base_aug_types`ì— ì •ì˜ëœ ê° ë³€í™˜ì— ëŒ€í•´ ë°˜ë³µí•©ë‹ˆë‹¤.
    for prefix, aug_function in base_aug_types:
        # ë³€í™˜ í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ì´ë¯¸ì§€ë¥¼ ë³€í™˜í•©ë‹ˆë‹¤.
        transformed_image = aug_function(image=image)['image']
        
        # ë³€í™˜ëœ ì´ë¯¸ì§€ì˜ ìƒˆë¡œìš´ IDë¥¼ ìƒì„±í•©ë‹ˆë‹¤.
        new_ID = prefix + ID
        
        # ìƒˆë¡œìš´ IDì™€ íƒ€ê²Ÿ ì •ë³´ë¥¼ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€í•©ë‹ˆë‹¤.
        ids.append(new_ID)
        targets.append(target)
        
        # ë³€í™˜ëœ ì´ë¯¸ì§€ë¥¼ ì €ì¥í•©ë‹ˆë‹¤.
        Image.fromarray(transformed_image).save(os.path.join(test_aug_image_path, new_ID))
        
# ë³€í™˜ëœ ì´ë¯¸ì§€ì˜ IDì™€ íƒ€ê²Ÿ ì •ë³´ë¥¼ ë‹´ì€ DataFrameì„ ìƒì„±í•©ë‹ˆë‹¤.
aug_data = {
    'ID': ids,
    'target': targets
}
aug_data = pd.DataFrame(aug_data)    

# ì›ë³¸ DataFrame `train_kr_df`ì™€ ì¦ê°•ëœ ë°ì´í„° `aug_data`ë¥¼ ê²°í•©í•˜ì—¬ ìƒˆë¡œìš´ DataFrame `df`ë¥¼ ë§Œë“­ë‹ˆë‹¤.
df = pd.concat([test_df, aug_data])
```

```python
df.to_csv(pre_path + 'data/test_aug.csv', index=False)
```
